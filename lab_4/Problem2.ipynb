```python
# mount folder from the network
# mount /mnt/nnds

path = "/mnt/nnds/tankerkoenig-data/prices/2022/01/2022-01-01-prices.csv"

df = pd.read_csv(path)
```

```python
df.info()
```

```python
e10_data = df[['date','e10']]
e10_data.query("e10 > 0", inplace=True)
```

```python
# Plotting the evolution of E10 prices
plt.figure(figsize=(20, 6))
plt.plot(e10_data['date'], e10_data['e10'])
plt.title('Evolution of E10 Prices')
plt.xlabel('Date')
plt.ylabel('Price (in Euros)')

plt.show()
```
