Thanks to matplotlib, Python is great for the visualization of functions and data.

a) Implement your favourite function f(x) as Python function f(x).

b) Create a list of values x using list comprehension or the range(start,stop,step) function.

c) Using list comprehension, create a list y that contains the functional values y = f(x).

d) Import matplotlib and create a plot of your function. Customize your plot using methods such as:

– plt.title(string) to set a title

– plt.xlabel(string) and plt.ylabel(string) to label your axes

– plt.xlim(lower,upper) and plt.ylim(lower,upper) to set the x and y range

– plt.grid(Boolean) to toggle on or off the grid

– plt.xticks(lst) and plt.yticks(lst) to specify tick locations


e) Use plt.savefig(strFileName) to save your plot as pdf.


```python
import numpy as np
import matplotlib.pyplot as plt # d)

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def f(x): # a)
    x = np.deg2rad(x)  # Convert degrees to radians
    result = 0
    for i in range(10):  # Calculate up to the 10th term
        term = ((-1) ** i) * (x ** (2 * i + 1)) / factorial(2 * i + 1)
        result += term
    return result


x_values = [i for i in range(361)] # b)
y_values = [f(x) for x in x_values] # c)


plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="y = sin(x)", color="blue")
plt.title("Plot of y = sin(x) using Taylor Series")
plt.xlabel("x (in degrees)")
plt.ylabel("y = sin(x)")
plt.xlim(0, 360)
plt.ylim(-1, 1)
plt.grid(True)
plt.xticks([i for i in range(0, 361, 45)])
plt.yticks([-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2])
plt.legend()



plt.savefig("sine_curve.pdf") # e)
plt.show()
```
